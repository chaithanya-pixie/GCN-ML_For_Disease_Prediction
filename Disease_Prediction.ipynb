```python
import pandas as pd

# Load the training and testing datasets
train_data = pd.read_csv('training_data.csv')
test_data = pd.read_csv('test_data.csv')

# Check the structure of the dataset
print(train_data.head())
print(test_data.head())
```

```python
print(train_data.columns)
print(test_data.columns)
```

```python
train_data = train_data.loc[:, ~train_data.columns.str.contains('Unnamed')]
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
train_data['prognosis'] = encoder.fit_transform(train_data['prognosis'])
test_data['prognosis'] = encoder.transform(test_data['prognosis'])
```

```python
print(train_data.head())
print(test_data.head())
```

```python
X_train = train_data.drop(columns=['prognosis'])
y_train = train_data['prognosis']

X_test = test_data.drop(columns=['prognosis'])
y_test = test_data['prognosis']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train model
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

```python
#Goal: Load the dataset and check its structure before converting it into a graph.

import pandas as pd #manipulation ,data-cleaning,exploration,integration,storage(csv)

# Load the dataset (replace with the correct filename)
df = pd.read_csv("training_data.csv")

# Display the first few rows
print(" First 5 rows of the dataset:")
print(df.head())

# Show column names
print("\n Column names in the dataset:")
print(df.columns)

# Check for missing values
print("\n Missing values in each column:")
print(df.isnull().sum())
```

```python
''' We have an extra column (Unnamed: 133) that is completely empty (4920 missing values).
We only need symptoms and diseases (not empty columns).'''

# Drop the unwanted column
df = df.drop(columns=["Unnamed: 133"], errors="ignore")

# Verify that it is removed
print(" Columns after dropping unnecessary ones:")
print(df.columns)
```

```python
'''Create a graph where symptoms and diseases are nodes
   Connect symptoms to diseases if they appear in a patient record'''

import networkx as nx

# Identify symptoms (all columns except 'prognosis')
symptom_columns = df.columns[:-1]  
disease_column = "prognosis"

# Create an empty graph
G = nx.Graph()

# Add Symptoms as nodes
for symptom in symptom_columns:
    G.add_node(symptom, type="symptom")

# Add Diseases as nodes (formatted properly)
for disease in df[disease_column].unique():
    formatted_disease = disease.strip().replace(" ", "_")  # Replace spaces with underscores
    G.add_node(formatted_disease, type="disease")

# Add Edges (Symptoms  Disease)
for _, row in df.iterrows():
    formatted_disease = row[disease_column].strip().replace(" ", "_")
    for symptom in symptom_columns:
        if row[symptom] == 1:  # If symptom is present
            G.add_edge(symptom, formatted_disease)

# Print graph details
print(f" Graph created with {G.number_of_nodes()} nodes and {G.number_of_edges()} edges")

# Show first few edges
print(" Sample edges:", list(G.edges)[:10])
```

```python
# Save the graph in .dat format 
#this will allow us to reload the graph anytime without rebuilding it from the dataset.
with open("graph_train_data.dat", "w") as f:
    for edge in G.edges():
        f.write(f"{edge[0]} {edge[1]}\n")

print(" Graph saved as graph_train_data.dat")
```

```python
''' Before we move to predictions, let's make sure the .dat file is correct by reloading the graph and checking its structure. '''

# Load the graph from the saved .dat file
G_loaded = nx.Graph()

with open("graph_train_data.dat", "r") as f:
    for line in f:
        node1, node2 = line.strip().split(maxsplit=1)  # Ensure correct splitting
        G_loaded.add_edge(node1, node2)

# Print graph details
print(f" Reloaded graph has {G_loaded.number_of_nodes()} nodes and {G_loaded.number_of_edges()} edges")

# Show first few edges
print(" Sample edges from reloaded graph:", list(G_loaded.edges)[:10])
```

```python
#let's visualize a small part of the graph to confirm the connections.

import matplotlib.pyplot as plt

# Select a small subgraph (first 20 nodes) for visualization
subgraph_nodes = list(G_loaded.nodes)[:20]
subgraph = G_loaded.subgraph(subgraph_nodes)

# Set figure size
plt.figure(figsize=(12, 8))

# Draw the graph
nx.draw(subgraph, with_labels=True, node_size=500, font_size=10, edge_color="gray")

# Show the plot
plt.title("Bipartite Graph Visualization (Sample)")
plt.show()
```

```python
#Plot the full graph 

import matplotlib.pyplot as plt
from networkx.algorithms import bipartite

# Identify symptom & disease nodes
symptoms = {node for node, data in G_loaded.nodes(data=True) if node in symptom_columns}
diseases = set(G_loaded) - symptoms

# Ensure bipartite layout
pos = nx.bipartite_layout(G_loaded, symptoms)

# Plot the graph
plt.figure(figsize=(15, 10))
nx.draw(G_loaded, pos, with_labels=True, node_size=300, font_size=8, edge_color="gray")

plt.title("Bipartite Graph of Symptoms and Diseases")
plt.show()
```

```python
''' Load the test dataset (test_data.csv)
    Predict diseases based on symptoms using the graph '''

# Load test dataset
test_df = pd.read_csv("test_data.csv")  # Replace with actual test file path

# Identify symptom columns (all except prognosis)
symptom_columns = test_df.columns[:-1]

# Extract symptoms where value == 1 for each patient
test_df["Symptoms"] = test_df[symptom_columns].apply(lambda row: row[row == 1].index.tolist(), axis=1)

# Display first few rows
print(" First 5 rows of test data with extracted symptoms:")
print(test_df[["Symptoms", "prognosis"]].head())
```

```python
'''  Use the graph to find the most connected disease for each set of symptoms
     Store the predicted disease in a new column
     Compare predicted diseases with actual diseases

 '''


# Function to predict disease based on symptoms using the graph
def predict_disease(symptoms):
    disease_count = {}  # Dictionary to count occurrences of diseases

    for symptom in symptoms:
        if symptom in G_loaded:
            connected_diseases = list(G_loaded.neighbors(symptom))  # Get diseases linked to symptom
            
            # Count occurrences of each disease
            for disease in connected_diseases:
                disease_count[disease] = disease_count.get(disease, 0) + 1

    # Return the disease with the highest occurrence
    return max(disease_count, key=disease_count.get) if disease_count else "Unknown"

# Apply the prediction function to test data
test_df["Predicted"] = test_df["Symptoms"].apply(predict_disease)

# Show actual vs predicted diseases
print(" First 5 rows of actual vs predicted diseases:")
print(test_df[["prognosis", "Predicted"]].head())
```

```python
''' To check how well our model is performing, we will:
    Compare actual vs predicted diseases
    Calculate accuracy, precision, recall, and F1-score '''

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Convert actual and predicted labels to lists
y_true = test_df["prognosis"].str.replace(" ", "_")  # Ensure format matches graph
y_pred = test_df["Predicted"]

# Accuracy
accuracy = accuracy_score(y_true, y_pred)

# Precision, Recall, and F1-Score (for multi-class classification)
precision = precision_score(y_true, y_pred, average='weighted', zero_division=1)
recall = recall_score(y_true, y_pred, average='weighted', zero_division=1)
f1 = f1_score(y_true, y_pred, average='weighted')

# Display results
print(f" Accuracy: {accuracy:.2f}")
print(f" Precision: {precision:.2f}")
print(f" Recall: {recall:.2f}")
print(f" F1 Score: {f1:.2f}")
```
